## Data Cleaning

```python
#Creating copy of data frame
fifa = data.copy()
fifa.sample(10)
fifa.info()
```

### Club Column
#### Player's Club

```python
fifa['Club'].dtype
```

```python
fifa['Club'].unique()
```

```python
#Removing leading spaces
fifa['Club'] = fifa['Club'].str.strip()
fifa['Club'].unique()
```

### Contract Column
#### Player's Contract with Club

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
# Defining needed function
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year + start_year
    return start_date, end_date, contract_length

# Apply function to Contract column and create new columns
new_cols = ['Contract Start', 'Contract End', 'Contract_length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
fifa.sample(10)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract_length(years)']].sample(5)
```

```python
# Contract Categories
def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'

# Add contract status Column
fifa.insert(fifa.columns.get_loc('Contract_length(years)')+1, 'Contract Status', fifa['Contract'].apply(categorize_contract_status))
fifa.sample(5)
```

```python
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract_length(years)', 'Contract Status']].sample(12)
```

### Height Column
#### Player's Height in centimeters

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
#Defining function to have one height measure
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)

# Apply function to height column
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns={'Height':"Height(cm)"})
fifa.sample(3)
```

### Weight Column
#### Player's weight in kilograms

```python
fifa['Weight'].dtype
```

```python
fifa['Weight'].unique()
```

```python
#Defining function to have one weight measure
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.05)
    
#Apply function to weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
fifa = fifa.rename(columns={"Weight": "Weight(kg)"})
fifa.sample(5)
```
